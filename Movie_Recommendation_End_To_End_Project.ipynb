## Data Preprocessing

Now we convert                                                                                                              
[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]                                                                                                       
This format to this                                                                                                             
['Action','Adventure','Fantacy','Scifi'....]                                                                                 
So for that atfirst we convert it using a python module ast.litral_eval to convert string of list to list                    

```python
import ast
def convert(obj):
    l = []
    for i in ast.literal_eval(obj):
        l.append(i['name'])
        return l
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies.head(3)
```

```python
movies['keywords'].value_counts()
```

### Now we will do same things on keywords

```python
movies['keywords'] = movies['keywords'].apply(convert)
movies.head()
```

```python
import ast
def convert2(obj):
    l = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter !=3:
            l.append(i['name'])
            counter+=1
        else:
            break
        return l
```

```python
movies['cast'] = movies['cast'].apply(convert2)
movies.head()
```

```python
## Now we fetch director name form crew column

def fetch_obj(obj):
    l = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            l.append(i['name'])
            break
    return l
```

```python
movies['crew'] = movies['crew'].apply(fetch_obj)
movies.head()
```

```python
movies['overview'][0]
```

### Now we just convert overview column into list so we can concatinate it

```python
movies.dropna(inplace=True)
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

### Now we just remove the spaces between names because to perform better performance

```python
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","")for i in x])
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","")for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","")for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","")for i in x])
```

```python
movies.head()
```

## Now we concat 'keywords','genres','cast','crew','overview' in one column

```python
movies['tags'] = movies['genres'] + movies['keywords'] + movies['overview'] + movies['cast'] + movies['crew']
movies.head()
```

```python
new_df = movies[['movie_id','title','tags']]
new_df.head()
```

### Converting list into strings

```python
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df['tags'][0]
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
new_df.head()
```

## Now we should do TextVectorization on tags column
Machine learning algorithms most often take numeric feature vectors as input. Thus, when working with text documents, we need a way to convert each document into a numeric vector. This process is known as text vectorization. One common vectorization approach is to represent each document as a vector of word counts.                                                        
In this use Bag of Words (BoW) Term Frequency method

Before TextVectorization atfirst we will do:
## Now we perform Stemmer from nltk library
Stemmers pick up the inclusion of a core or stem word within a longer word. For example, a stemming algorithm might look at a word like “planning,” and correctly recognize that the root word or stem word is “plan.” This can be a helpful element of something that parses raw text for analysis

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
new_df['tags'][0]
```

### Now from here we use TextVectorization

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
vectors
```

```python
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Batman Begins')
```
